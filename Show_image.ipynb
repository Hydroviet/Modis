```python
import matplotlib.pyplot as plt
import numpy as np
import rasterio as rio
%matplotlib inline
```

```python
def findNIR(i, year, day):
    import os
    prefix = 'MOD13Q1/{}/{}/{}{}'.format(i, year, year, str(day).zfill(3))
    listRaster = os.listdir(prefix)
    NIR = list(filter(lambda x: 'NIR' in x, listRaster))
    return prefix + '/' + NIR[0]
```

```python
year = 2007
day = 49
fig, axes = plt.subplots(60//4 + 1, 4, figsize=(40,100))
for i in range(0, 61, 4):
    for j in range(4):
        with rio.open(findNIR(i + j, year, day)) as src:
            NIR = src.read(1)
        rowIndex = (i)//4
        #print(rowIndex)
        axes[rowIndex, j].imshow(NIR)
        axes[rowIndex, j].set_title('{} - {}'.format(i + j, str(NIR.shape)))
```
